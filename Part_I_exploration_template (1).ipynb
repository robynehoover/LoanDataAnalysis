## Multivariate Exploration

```python
g = sb.FacetGrid(data = df, hue = 'IncomeRange', size = 5, palette = 'Blues')
g.map(plt.scatter, 'BorrowerRate', 'BorrowerAPR')
g.add_legend()
plt.title('APR vs. Interest Rate by Income Range')
```

```python
plt.figure(figsize=[20,20])
cat_means = df.groupby(['IncomeRange', 'BorrowerState']).mean()['BorrowerRate']
cat_means = cat_means.reset_index(name = 'Interest Rate')
cat_means = cat_means.pivot(index = 'BorrowerState', columns = 'IncomeRange',
                            values = 'Interest Rate')
sb.heatmap(cat_means, annot = True, fmt = '.3f',
           cbar_kws = {'label' : 'Interest Rate'})
plt.title('How Income Range and State affect Interest Rates')
plt.xlabel('Income Range ($)')
```

Borrowers with lower income ranges and unemployed borrowers tend to have higher interest rates. Interest rates do not seem to differ much by state.

```python
plt.figure(figsize=[10,10])
cat_means = df.groupby(['IncomeRange', 'CreditScoreRangeUpper']).mean()['BorrowerRate']
cat_means = cat_means.reset_index(name = 'Interest Rate')
cat_means = cat_means.pivot(index = 'CreditScoreRangeUpper', columns = 'IncomeRange',
                            values = 'Interest Rate')
sb.heatmap(cat_means, annot = True, fmt = '.3f',
           cbar_kws = {'label' : 'Interest Rate'})
plt.title('How Credit Score and Income affect Interest Rate')
plt.xlabel('Income Range ($)')
```

A higher income and higher credit score tends to lead to a lower interest rate. This figure shows that having a higher credit score is more important than having a higher income though when trying to obtain a lower interest rate.

```python
plt.figure(figsize=[10,10])
cat_means = df.groupby(['Term', 'CreditScoreRangeUpper']).mean()['BorrowerRate']
cat_means = cat_means.reset_index(name = 'Interest Rate')
cat_means = cat_means.pivot(index = 'CreditScoreRangeUpper', columns = 'Term',
                            values = 'Interest Rate')
sb.heatmap(cat_means, annot = True, fmt = '.3f',
           cbar_kws = {'label' : 'Interest Rate'})
plt.title('How Credit Score and Loan Term affect Interest Rate')
plt.xlabel('Loan Term (Months)')
```

A higher credit score and a loan term of 12 months tend to have the lowest interest rates.

```python
plt.figure(figsize=[15,10])
cat_means = df.groupby(['ListingCategory (numeric)', 'CreditScoreRangeUpper']).mean()['BorrowerRate']
cat_means = cat_means.reset_index(name = 'Interest Rate')
cat_means = cat_means.pivot(index = 'CreditScoreRangeUpper', columns = 'ListingCategory (numeric)',
                            values = 'Interest Rate')
sb.heatmap(cat_means, annot = True, fmt = '.3f',
           cbar_kws = {'label' : 'Interest Rate'})
plt.title('How Loan Type and Credit Score affect Interest Rate')
plt.xlabel('Loan Type (numeric)')
```

Certain loan types have higher interest rates than others, but having a higher credit score can help reduce the interest rate.
